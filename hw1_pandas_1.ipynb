```python
# –ó–∞–¥–∞—á–∞ 1: –°–≥–µ–Ω–µ—Ä–∏—Ä–æ–≤–∞—Ç—å –º–∞—Å—Å–∏–≤ –∏–∑ —Å–ª—É—á–∞–π–Ω—ã—Ö –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö —á–∏—Å–µ–ª –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω–æ–π –¥–ª–∏–Ω—ã. –í—ã–≤–µ—Å—Ç–∏ —ç–ª–µ–º–µ–Ω—Ç—ã, –∫–æ—Ç–æ—Ä—ã–µ 
# –±–æ–ª—å—à–µ —Å—Ä–µ–¥–Ω–µ–≥–æ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–≥–æ.

arr = np.random.uniform(0, 100, 20)  # 20 —Å–ª—É—á–∞–π–Ω—ã—Ö –¥—Ä–æ–±–Ω—ã—Ö —á–∏—Å–µ–ª –æ—Ç 0 –¥–æ 100
mean = np.mean(arr)  # —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ
more_than_mean = [i for i in arr if i>mean]  # —Å–æ–∑–¥–∞–µ–º –ª–∏—Å—Ç –∫–æ–º–ø—Ä–µ—Ö–µ–Ω—à–Ω|–ª–∏–±–æ –∏—Å–ø–æ–ª—å–∑—É—è numpy: more_than_mean = arr[arr > mean]
print(f"–ú–∞—Å—Å–∏–≤: {arr} \n")
print(f"–°—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ: {mean} \n")
print(f"–≠–ª–µ–º–µ–Ω—Ç—ã –±–æ–ª—å—à–µ —Å—Ä–µ–¥–Ω–µ–≥–æ: {more_than_mean}")
```

```python
# –ó–∞–¥–∞—á–∞ 2: –°–≥–µ–Ω–µ—Ä–∏—Ä–æ–≤–∞—Ç—å –º–∞—Å—Å–∏–≤ –∏–∑ 20 —Å–ª—É—á–∞–π–Ω—ã—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤, —ç–ª–µ–º–µ–Ω—Ç—ã –∫–æ—Ç–æ—Ä–æ–≥–æ –ª–µ–∂–∞—Ç –≤ –∏–Ω—Ç–µ—Ä–≤–∞–ª–µ [0, 10].
# –û–ø—Ä–µ–¥–µ–ª–∏—Ç—å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤, –æ—Ç–ª–∏—á–∞—é—â–∏—Ö—Å—è –æ—Ç –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–≥–æ –Ω–∞ 5.
# –ù–∞–π—Ç–∏ —Å—Ä–µ–¥–Ω–µ–µ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ –≤—Å–µ—Ö —á–µ—Ç–Ω—ã—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –º–∞—Å—Å–∏–≤–∞, —Å—Ç–æ—è—â–∏—Ö –Ω–∞ –Ω–µ—á–µ—Ç–Ω—ã—Ö –º–µ—Å—Ç–∞—Ö.

arr_20 = np.random.randint(0, 11, 20)

min_elem = np.min(arr_20)
diff_5 = np.sum(np.abs(arr_20 - min_elem) == 5)

even_numbers_at_odd = [arr_20[i] for i in range(1, len(arr_20), 2) if arr_20[i] % 2 == 0]
mean_even_odd = np.mean(even_numbers_at_odd) # if even_numbers_at_odd else None

print(f"–°–ª—É—á–∞–π–Ω—ã–π –º–∞—Å—Å–∏–≤: {arr_20}")
print(f"–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤, –æ—Ç–ª–∏—á–∞—é—â–∏—Ö—Å—è –æ—Ç –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–≥–æ –Ω–∞ 5: {diff_5}")
print(f"–°—Ä–µ–¥–Ω–µ–µ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ —á–µ—Ç–Ω—ã—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –Ω–∞ –Ω–µ—á–µ—Ç–Ω—ã—Ö –º–µ—Å—Ç–∞—Ö: {mean_even_odd:.2f}")
```

```python
# –ó–∞–¥–∞—á–∞ 3: –°–≥–µ–Ω–µ—Ä–∏—Ä–æ–≤–∞—Ç—å –º–∞—Ç—Ä–∏—Ü—É A —Ä–∞–∑–º–µ—Ä–æ–º 5 –Ω–∞ 5, —ç–ª–µ–º–µ–Ω—Ç—ã –∫–æ—Ç–æ—Ä–æ–π –ª–µ–∂–∞—Ç –≤ –∏–Ω—Ç–µ—Ä–≤–∞–ª–µ [1, 10].
# –í–æ–∑–≤–µ—Å—Ç–∏ –≤ –∫–≤–∞–¥—Ä–∞—Ç –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π —ç–ª–µ–º–µ–Ω—Ç –º–∞—Ç—Ä–∏—Ü—ã.
# –í—Å–µ —á–µ—Ç–Ω—ã–µ —á–∏—Å–ª–∞ –∑–∞–º–µ–Ω–∏—Ç—å –Ω–∞ 0.

matrix_A = np.random.randint(1, 11, (5, 5))
matrix_A
```

```python
max_value = np.max(matrix_A)
matrix_A[matrix_A == max_value] = max_value ** 2  # –í–æ–∑–≤–µ–¥–µ–Ω–∏–µ –≤ –∫–≤–∞–¥—Ä–∞—Ç –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞
matrix_A[matrix_A % 2 == 0] = 0  # –ó–∞–º–µ–Ω–∞ —á–µ—Ç–Ω—ã—Ö —á–∏—Å–µ–ª –Ω–∞ 0

print("–ú–æ–¥–∏—Ñ–∏—Ü–∏—Ä–æ–≤–∞–Ω–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞ A:")
matrix_A
```

```python
# –ó–∞–¥–∞—á–∞ 4: –°–≥–µ–Ω–µ—Ä–∏—Ä–æ–≤–∞—Ç—å –º–∞—Ç—Ä–∏—Ü—É B —Ä–∞–∑–º–µ—Ä–æ–º 5 –Ω–∞ 5, —ç–ª–µ–º–µ–Ω—Ç—ã –∫–æ—Ç–æ—Ä–æ–π –ª–µ–∂–∞—Ç –≤ –∏–Ω—Ç–µ—Ä–≤–∞–ª–µ [-5, 5]. 
# –û—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–µ —ç–ª–µ–º–µ–Ω—Ç—ã –º–∞—Ç—Ä–∏—Ü—ã –∑–∞–º–µ–Ω–∏—Ç—å –Ω–∞ —á–∏—Å–ª–æ –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤ —Å—Ç—Ä–æ–∫–µ, –≤ –∫–æ—Ç–æ—Ä–æ–π –æ–Ω–∏ –Ω–∞—Ö–æ–¥—è—Ç—Å—è.
matrix_B = np.random.randint(-5, 6, (5, 5))
matrix_B
```

```python
for i in range(matrix_B.shape[0]):
    negative_count = np.sum(matrix_B[i] < 0)
    matrix_B[i][matrix_B[i] < 0] = negative_count  # p–∞–º–µ–Ω–∞ –æ—Ç—Ä–∏—Ü —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –Ω–∞ –∏—Ö –∫–æ–ª-–≤–æ –≤ —Å—Ç—Ä–æ–∫–µ

print("–ú–æ–¥–∏—Ñ–∏—Ü–∏—Ä–æ–≤–∞–Ω–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞ B:")
print(matrix_B)
```

# –ó–∞–¥–∞–Ω–∏–µ 3: –°–∞–º—ã–µ –ø–æ–ø—É–ª—è—Ä–Ω—ã–µ –∏–º–µ–Ω–∞ —Å—Ä–µ–¥–∏ –º—É–∂—á–∏–Ω –∏ –∂–µ–Ω—â–∏–Ω

```python
import re

# a—É–Ω–∫—Ü–∏—è –¥–ª—è –∏–∑–≤–ª–µ—á–µ–Ω–∏—è –∏–º–µ–Ω–∏
def get_first_name(name):
    match = re.search(r'\(([^)]+)\)', name)  # –ø—Ä–æ–≤–µ—Ä—è–µ–º, –µ—Å—Ç—å –ª–∏ –∏–º—è –≤ —Å–∫–æ–±–∫–∞—Ö
    if match:
        return match.group(1).split()[0]  # –±–µ—Ä—ë–º –ø–µ—Ä–≤–æ–µ —Å–ª–æ–≤–æ –∏–∑ —Å–∫–æ–±–æ–∫ (–∂–µ–Ω—Å–∫–æ–µ –∏–º—è)
    else:
        return name.split('.')[1].strip().split()[0]  # –±–µ—Ä—ë–º –ø–µ—Ä–≤–æ–µ —Å–ª–æ–≤–æ –ø–æ—Å–ª–µ —Ç–∏—Ç—É–ª–∞
```

```python
# –Ω–æ–≤—ã–π —Å—Ç–æ–ª–±–µ—Ü —Å –∏–º–µ–Ω–∞–º–∏
df['first_name'] = df['name'].apply(get_first_name)
df.sample(3)
```

```python
# –ø–ø—Ä–æ–≤–µ—Ä—è–µ–º, —Å–∞–º—ã–µ –ø–æ–ø—É–ª—è—Ä–Ω—ã–µ –∏–º–µ–Ω–∞
most_common_male_name = df[df['sex'] == 'male']['first_name'].mode()[0]
most_common_female_name = df[df['sex'] == 'female']['first_name'].mode()[0]

print(f"–°–∞–º–æ–µ –ø–æ–ø—É–ª—è—Ä–Ω–æ–µ –∏–º—è —Å—Ä–µ–¥–∏ –º—É–∂—á–∏–Ω: {most_common_male_name}")
print(f"–°–∞–º–æ–µ –ø–æ–ø—É–ª—è—Ä–Ω–æ–µ –∏–º—è —Å—Ä–µ–¥–∏ –∂–µ–Ω—â–∏–Ω: {most_common_female_name}")
```

# –ó–∞–¥–∞–Ω–∏–µ 6: –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ crosstab –∏ pivot_table

### - *crosstab()*

–°–æ–∑–¥–∞—ë—Ç —Ç–∞–±–ª–∏—Ü—É —á–∞—Å—Ç–æ—Ç (—Å—á—ë—Ç–∞) –∑–Ω–∞—á–µ–Ω–∏–π –º–µ–∂–¥—É –¥–≤—É–º—è —Å—Ç–æ–ª–±—Ü–∞–º–∏.
–ò—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è, –∫–æ–≥–¥–∞ –Ω—É–∂–Ω–æ –ø–æ—Å—á–∏—Ç–∞—Ç—å –≤–∑–∞–∏–º–æ—Å–≤—è–∑—å –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö.

### - pivot_table()

–ì–∏–±—á–µ, —á–µ–º crosstab(), –ø–æ–∑–≤–æ–ª—è–µ—Ç –∞–≥—Ä–µ–≥–∏—Ä–æ–≤–∞—Ç—å –¥–∞–Ω–Ω—ã–µ —Ä–∞–∑–Ω—ã–º–∏ —Å–ø–æ—Å–æ–±–∞–º–∏.
–ò—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è, –∫–æ–≥–¥–∞ –Ω—É–∂–Ω–æ –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∫–∞ + –∞–≥—Ä–µ–≥–∏—Ä–æ–≤–∞–Ω–∏–µ (–Ω–∞–ø—Ä–∏–º–µ—Ä, —Å—Ä–µ–¥–Ω–µ–µ, —Å—É–º–º–∞, –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ).

```python
# –î–æ–ª—è –≤—ã–∂–∏–≤—à–∏—Ö –ø–æ –ø–æ–ª—É
survival_by_gender = pd.crosstab(df['sex'], df['survived'], normalize='index')
survival_by_gender
```

```python
# –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–∞—Å—Å–∞–∂–∏—Ä–æ–≤ –ø–æ –∫–ª–∞—Å—Å–∞–º –∏ –ø–æ–ª—É
pclass_distribution_pivot = df.pivot_table(index='sex', columns='pclass', aggfunc='size', fill_value=0)
pclass_distribution_pivot
```

```python
# –î–æ–ª—è –≤—ã–∂–∏–≤—à–∏—Ö –≤ —Ä–∞–∑–Ω—ã—Ö –≤–æ–∑—Ä–∞—Å—Ç–Ω—ã—Ö –≥—Ä—É–ø–ø–∞—Ö
survival_by_age_group = pd.crosstab(df['age_group'], df['survived'], normalize='index')
survival_by_age_group
```
